## 311 Service Requests - Vacant and Abandoned Buildings Reported - Historical

```python
client = Socrata("data.cityofchicago.org", None)
```

```python
abandon = client.get("7nii-7srd", limit=100000)
abandon = pd.DataFrame.from_records(abandon)
```

```python
len(abandon)
```

```python
abandon.head()
```

```python
buildings=[[i, r['service_request_number'], r['date_service_request_was_received'][:4], float(r['longitude']), float(r['latitude']), str(r['longitude'])+','+str(r['latitude'])]for i, r in abandon.iterrows()]
```

```python
buildings_df = pd.DataFrame(buildings, columns=['id','service_request_number', 'year', 'longitude', 'latitude', 'unique'])
buildings_df.head()
```

```python
geometry = [Point(xy) for xy in zip(buildings_df.longitude, buildings_df.latitude)]
crs = {'init': 'epsg:4326'}
geo_abandon = gpd.GeoDataFrame(buildings_df, crs=crs, geometry=geometry)
geo_abandon.to_file("clean_data/geo_abandon.shp")
```

## Small business fund

```python
buss_fund = util.get_data("jp7n-tgmf", 1000000)
for c in buss_fund.columns:
    print(c)
len(buss_fund)
```

```python
buss_fund.head()
```

```python
buss_fund_list=[[i, int(r['completion_date'][:4]), r['company'], r['address']]for i, r in buss_fund.iterrows()]
buss_fund_df = pd.DataFrame(buss_fund_list, columns=['id', 'year', 'company', 'address'])
buss_fund_df = buss_fund_df[(buss_fund_df['year']<2018)&(buss_fund_df['year']>2011)]
buss_fund_df.head()
```

```python
buss_fund_df['year'].unique()
```

```python
client = googlemaps.Client(key=creds.google_maps_key)
complete_data = []
```

```python
for i,r in buss_fund_df.iterrows():
    try:
        info = client.geocode(r['address']+'chicago')[0]
        complete_data.append([r['id'], r['year'], info['geometry']['location']['lat'], info['geometry']['location']['lng']])
    except IndexError:
        pass
```

```python
complete_data
```

## Building violations
